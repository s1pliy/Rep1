# **Створення таблиці у Python**




Ця програма використовує клас PrettyTable з бібліотеки prettytable для створення таблиці зі стовпцями часу та діяльності. Потім він додає рядки для різних дій протягом дня, замінюючи комірку, що відповідає 12:00, малюнком кота. Запустіть цю програму, і ви побачите кота в таблиці розкладу у вашому терміналі.


```python

from prettytable import PrettyTable

def draw_cat():
    cat = r"""
 /\_/\
( o.o )
 >^_^<
"""
    return cat

def create_schedule():
    schedule = PrettyTable()
    schedule.field_names = ["Time", "Activity"]
    schedule.add_row(["9:00 AM", "Meeting"])
    schedule.add_row(["10:00 AM", "Work"])
    schedule.add_row(["11:00 AM", "Lunch"])
    # Replace the placeholder cell with the cat drawing
    cat_row = ["12:00 PM", draw_cat()]
    schedule.add_row(cat_row)
    schedule.add_row(["1:00 PM", "Work"])
    schedule.add_row(["2:00 PM", "Coding"])
    schedule.add_row(["3:00 PM", "Exercise"])
    schedule.add_row(["4:00 PM", "Reading"])
    schedule.add_row(["5:00 PM", "Relaxing"])
    print(schedule)

# Call the function to create and display the schedule
create_schedule()
```

# **Графік синусоїда в Python**

Цей код імпортує бібліотеку NumPy для створення даних і бібліотеку Matplotlib для створення простого графіка синусоїди. Він створює графік синусоїди та відображає його. Коли ви запускаєте цей код у блокноті Colab, він створить графік і відобразить його безпосередньо в інтерфейсі блокнота.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave')
plt.title('Sine Wave Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

# **Калкулятор в Python**

Вам буде запропоновано вибрати дію (додавання, віднімання, множення або ділення), а потім ввести два числа. Після введення чисел він виконає вибрану операцію та відобразить результат.

```python
# Function to add two numbers
def add(x, y):
    return x + y

# Function to subtract two numbers
def subtract(x, y):
    return x - y

# Function to multiply two numbers
def multiply(x, y):
    return x * y

# Function to divide two numbers
def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Division by zero is not allowed."

# Main function to take user input and perform calculations
def calculator():
    print("Select operation:")
    print("1. Додавання")
    print("2. Віднімання")
    print("3. Множення")
    print("4. Ділення")

    choice = input("Enter choice (1/2/3/4): ")

    num1 = float(input("Введіть перше число: "))
    num2 = float(input("Введіть друге число: "))

    if choice == '1':
        print("Result: ", add(num1, num2))
    elif choice == '2':
        print("Result: ", subtract(num1, num2))
    elif choice == '3':
        print("Result: ", multiply(num1, num2))
    elif choice == '4':
        print("Result: ", divide(num1, num2))
    else:
        print("Помилка. Будь ласка, введіть один із запропонованих варіантів.")

# Call the calculator function
calculator()
```
